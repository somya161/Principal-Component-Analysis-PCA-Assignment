```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv("wine.csv")
wine
```

```python
Wine= wine.iloc[:,1:]
Wine
```

```python
# Converting data to numpy array
wine_ary=Wine.values
wine_ary
```

```python
# Normalizing the  numerical data
wine_norm=scale(wine_ary)
wine_norm
```

```python
# Applying PCA Fit Transform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
# PCA Components matrix or convariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="purple");
```

```python
# Final Dataframe
pca_df = pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
pca_df
```

```python
principal_component1 = pca_df['PC1']
principal_component2 = pca_df['PC2']
principal_component3 = pca_df['PC3']
```

```python
# Visualization of PCAs
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=pca_df);
```

```python
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Type');
```

```python
# Visualizing the results of the 3D PCA.
ax = plt.figure(figsize=(10,10)).gca(projection='3d')
plt.title('3D Principal Component Analysis (PCA)')
ax.scatter(
    xs=principal_component1, 
    ys=principal_component2, 
    zs=principal_component3, 
)
ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')
plt.show()
```

### Agglomerative Clustering

```python
# Importing libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# As we already have normalized data, create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
# Creating Clusters
HC_clusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
HC_clusters
```

```python
y = pd.DataFrame(HC_clusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
wine2=wine.copy()
wine2['clustersid'] = HC_clusters.labels_
wine2
```

```python
wine2.groupby('clustersid').agg(['mean']).reset_index()
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_score= silhouette_score(wine_norm, HC_clusters.labels_) #silhouette_score of AgglomerativeClustering
sil_score
```

### KMeans Clustering

```python
# importing libraries
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
# As we already have normalized data
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change
```

```python
# Using data normalized by MinMaxScaler
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
wine3=wine.copy()
wine3['clusters3id']=clusters3.labels_
wine3
```

```python
wine3.groupby('clusters3id').agg(['mean']).reset_index()
```
